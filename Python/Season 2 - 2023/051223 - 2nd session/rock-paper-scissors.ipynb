```python
import random as rand
import time

def delay(delayInSeconds):
    time.sleep(delayInSeconds)

def updateScore(player, point):
    player["score"] += point

def checkScore(player):
    return player["score"]

def showScore(userScore = 0, computerScore = 0):
    print(f"""
SCORE
You: {userScore["score"]}
Computer: {computerScore["score"]}
    """)

CHARACTERS = ["Fire", "Water", "Grass"]
ENVIRONMENT = [""]
USER_WON = [[1, 3], [2, 1], [3, 2]]
TIE = [[1, 1], [2, 2], [3, 3]]
# dont need COMPUTER_WON because we're going to use "else" instead
RESULTS = [
    ["You won!"],
    ["Computer won!"],
    ["It's a tie!"]
]

playing = True

WELCOME_TEXT = """
CHOOSE A CHARACTER
PRESS THE NUMBER FOR YOUR CHARACTER
1 - Fire
2 - Water
3 - Grass
"""

delay(1)

userScore = {"score": 0}
computerScore = {"score": 0}

while playing:
    # show "WELCOME_TEXT" to user
    print(WELCOME_TEXT)

    userChoice = int(input())
    computerChoice = rand.randint(1, 3)
    userAndcomputerChoice = [userChoice, computerChoice]

    delay(1)

    if userAndcomputerChoice in USER_WON:
        print(RESULTS[0][0])
        updateScore(userScore, 1)
        showScore(userScore, computerScore)
    elif userAndcomputerChoice in TIE:
        print(RESULTS[2][0])
        showScore(userScore, computerScore)
    else:
        print(RESULTS[1][0])
        updateScore(computerScore, 1)
        showScore(userScore, computerScore)

    if 10 in [checkScore(userScore), checkScore(computerScore)]:
        playing = False
```
