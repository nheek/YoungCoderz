```python
import random as rand
import time

# Constants for character choices and game results
CHARACTERS = ["Fire", "Water", "Grass"]
ENVIRONMENT = [""]
USER_WON = [[0, 2], [1, 0], [2, 1]]  # Adjusted for 0-based index
TIE = [[0, 0], [1, 1], [2, 2]]       # Adjusted for 0-based index
RESULTS = [
    ["You won!"],
    ["Computer won!"],
    ["It's a tie!"]
]
WINNING_SCORE = list(range(10, 15))

# Variable to control game loop
playing = True

# Welcome message for the user
WELCOME_TEXT = """
CHOOSE A CHARACTER
PRESS THE NUMBER FOR YOUR CHARACTER
0 - Fire
1 - Water
2 - Grass
"""

# Player scores and game state variables
userScore = {"score": 0}
computerScore = {"score": 0}

powerup = False
powerupPlayer = ""
powerupCounterPlayer = ""
powerupCounter = 0

immunity = False
immunityPlayer = ""

# Function to introduce a delay in the game
def delay(delayInSeconds):
    time.sleep(delayInSeconds)

# Function to update the score of a player
def updateScore(player, point):
    player["score"] += point

# Function to retrieve the current score of a player
def checkScore(player):
    return player["score"]

# Function to display the current scores
def showScore(userScore=0, computerScore=0):
    print(f"""
SCORE
You: {userScore}
Computer: {computerScore}
    """)

# Function to update the streak and activate power-up
def updateStreak(forWhichPlayer):
    global powerupCounterPlayer, powerupCounter, powerup, powerupPlayer
    if powerupCounterPlayer != forWhichPlayer:
        powerupCounterPlayer = forWhichPlayer
        powerupCounter = 1
    else:
        powerupCounter += 1

    if powerupCounter >= 3:
        print(f"{forWhichPlayer} has a 2x point multiplier!")
        powerupPlayer = forWhichPlayer
        powerup = True

# Function to check and apply power-ups
def checkPowerup(forWhichPlayer, scoreVar):
    global powerup, powerupPlayer
    if powerup and powerupPlayer == forWhichPlayer:
        updateScore(scoreVar, 2)
    else:
        updateScore(scoreVar, 1)

# Main game loop
while playing:
    # Show welcome text to the user
    print(WELCOME_TEXT)

    # User makes a choice
    userChoice = int(input())
    print(f"You chose {CHARACTERS[userChoice]}")

    # Countdown before computer makes a choice
    delay(1)
    print("2...")
    delay(1)
    print("1...")
    delay(1)
    
    # Computer makes a random choice
    computerChoice = rand.randint(0, 2)
    print(f"Computer chose {CHARACTERS[computerChoice]}\n")

    # Combine user and computer choices for comparison
    userAndcomputerChoice = [userChoice, computerChoice]

    delay(1)

    # Compare choices and update scores
    if userAndcomputerChoice in USER_WON:
        print(RESULTS[0][0])
        checkPowerup("user", userScore)
        showScore(userScore["score"], computerScore["score"])
        updateStreak("user")

    elif userAndcomputerChoice in TIE:
        print(RESULTS[2][0])
        showScore(userScore["score"], computerScore["score"])
    else:
        print(RESULTS[1][0])
        checkPowerup("computer", computerScore)
        showScore(userScore["score"], computerScore["score"])
        updateStreak("computer")

    # Check for a winner and end the game if a winning score is reached
    if any(score in WINNING_SCORE for score in [checkScore(userScore), checkScore(computerScore)]):
        winner = "User" if checkScore(userScore) > checkScore(computerScore) else "Computer"
        print(f"{winner} WON")
        playing = False

    print("--------------------------------")
```
